```python
# ============================================
# src/common/paths.ipynb
#
# PURPOSE
# - Standardize Volume-based path patterns used across ALL domains/pipelines.
#
# VOLUME PATH FORMAT
# /Volumes/<CATALOG>/<DOMAIN>/<ZONE>/<ENTITY>/
#
# Where:
# - DOMAIN is the UC schema that owns the volumes for that domain (e.g., maintenance, transport)
# - ZONE is one of: landing, checkpoints, schemas
# - ENTITY is the dataset name (e.g., truck_events)
#
# IMPORTANT NAMING RULE
# ============================================

CATALOG = "workspace"

# Standard zone folder names under each domain volume root
LANDING_ZONE = "landing"
CHECKPOINTS_ZONE = "checkpoints"
SCHEMAS_ZONE = "schemas"

# Root
VOLUME_ROOT = "/Volumes/{catalog}/{domain}"

# Zone base templates
LANDING_BASE = VOLUME_ROOT + f"/{LANDING_ZONE}"
CHECKPOINTS_BASE = VOLUME_ROOT + f"/{CHECKPOINTS_ZONE}"
SCHEMAS_BASE = VOLUME_ROOT + f"/{SCHEMAS_ZONE}"

# Entities
LANDING_PATH = LANDING_BASE + "/{entity}"
CHECKPOINT_PATH = CHECKPOINTS_BASE + "/{entity}"
SCHEMA_LOCATION = SCHEMAS_BASE + "/{entity}"
```
