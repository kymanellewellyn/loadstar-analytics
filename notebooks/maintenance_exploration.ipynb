```python
df = spark.read.text("/Volumes/loadstar/maintenance_raw/truck_maintenance_events_raw/events")
display(df)
```

```python
sample_df = spark.read.json("/Volumes/loadstar/maintenance_raw/truck_maintenance_events_raw/events")
schema = sample_df.schema
```

```python
from pyspark.sql.functions import from_json, col

parsed_df = (
    df
    .withColumn("json", from_json(col("value"), schema))
    .select("json.*")
)

display(parsed_df)
```
