```python
from src.common.config import CATALOG, DOMAINS

spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG}")
```

```python
for domain_name, domain_config in DOMAINS.items():
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{domain_config['raw_schema']}")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{domain_config['bronze_schema']}")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{domain_config['silver_schema']}")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{domain_config['gold_schema']}")

print(f"Catalog created or already exists: {CATALOG}")
print("Schemas created or already exist:")
```
